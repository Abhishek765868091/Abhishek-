```python
def is_prime(num: int) ->bool:
    """
       Args:
            take a lower and upper limit in integer value.
       Return:
             give all prime number in a integer value.
    """
    if not isinstance(num,int):
        raise TypeError("number must be integer")
    if not num>2:
        raise ValueError("number must be greater than 2")
    for i in range(2,int(num**.5)+1):
        if num%i==0:
            return False
    else:
        return True
if __name__=="__main__":
    try:
        prime_num=[]
        lower_limit=input("enter lower limit:")
        upper_limit=input("enter upper limit:")
        if not lower_limit.strip().isdigit():
            raise TypeError("lower input must be integer")
        if not upper_limit.strip().isdigit():
            raise TypeError("lower input must be integer")
        lower_limit=int(lower_limit)
        upper_limit=int(upper_limit)
        for i in range(lower_limit,upper_limit):
           if is_prime(i):
              print(f"{i} is a prime number")
              prime_num.append(i)
        print(prime_num)
        
    except TypeError as e:
        print("number must be integer",e)
    except ValueError as e:
        print(e)
    except Exception as e:
        print(e)
```
